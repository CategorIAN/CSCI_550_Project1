## Part 3.1: K-Means Algorithm (Ian)

### Preliminaries

I am using a small data set to test on. This data corresponds to soybean data.

```python
df = pd.read_csv("soybean_data.csv", index_col=0)
```

### Create Distance Matrix Function

```python
def distMatrix(df, centroids):
    df_mat, cent_mat = df.to_numpy(), centroids.to_numpy()
    return np.array([[np.linalg.norm(df_mat[i, :] - cent_mat[j, :]) for j in range(cent_mat.shape[0])] for i in range(df_mat.shape[0])])
```

### Assign New Clusters

```python
def cluster(dm):
    def f(i):
        return reduce(lambda x, y: x if x[0] <= y[0] else y, zip(dm[i, :], range(len(dm[i, :]))))[1]
    return f
```

### Bringing It All Together

```python
def kmeans(df, k, eps):
    @tail_recursive
    def go(currentClus, prevClus, r):
        print("Iteration Number {}".format(r))
        centroids = pd.DataFrame(df.to_dict(orient='series') | {
                                 "Cluster": currentClus}).groupby(by=["Cluster"]).mean()
        if prevClus is not None and ((currentClus != prevClus).sum() / len(currentClus)) < eps:
            return centroids
        else:
            dm = distMatrix(df, centroids)
            return go.tail_call(df.index.map(cluster(dm)), currentClus, r + 1)
    return go([randrange(k) for i in range(df.shape[0])], None, 0)
```

## Part 3.1: DBSCAN Algorithm (Chris)

### Implementation of DBSCAN Algorithm

Cluster labels of each data-point are assigned as follows: -1 for noise, 0 for border, 1, 2, 3, ... for core points.

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN as sklearnDBSCAN

def DBSCAN(data, minpts, epsilon):
    
    
    #find distance between two points
    def distance(x1, x2):
        return np.sqrt(np.sum((x1 - x2) ** 2))
    
    #for each data point, return the data indices that are within the epsilon range
    def range_query(data, idx):
        neighbors = []
        for i in range(len(data)):
            if distance(data[idx], data[i]) <= epsilon:
                neighbors.append(i)
        return neighbors

    #update and assign data point as either core, border or noise
    def update_cluster(data, point_index, cluster_label, neighbors, cluster):
        cluster[point_index] = cluster_label
        for i in neighbors:
            if cluster[i] is None:
                cluster[i] = cluster_label
                neighbor_neighbors = range_query(data, i)
                if len(neighbor_neighbors) >= minpts:
                    neighbors.extend(neighbor_neighbors)

                    
    #initilizing                 
    samples = len(data)
    cluster = [None] * samples
    cluster_label = 0
    
    #Skipping already labeled points
    for i in range(samples):
        if cluster[i] is not None:
            continue

        neighbors = range_query(data, i)
        if len(neighbors) < minpts:
            cluster[i] = -1  #Noise points
        else:
            update_cluster(data, i, cluster_label, neighbors, cluster)
            cluster_label += 1

    #Assigning border points to 0
    for i in range(samples):
        if cluster[i] is None:
            for neighbor in range_query(data, i):
                if cluster[neighbor] != -1:
                    cluster[i] = 0 
                    break

    #Collecting clusters
    unique_labels = set(cluster)
    clusters = []
    for label in unique_labels:
        if label != -1:
            clusters.append([data[i] for i in range(samples) if cluster[i] == label])

    return clusters, cluster

#Testing with make_moons from sklearn:

TEST,_= make_moons(n_samples=100, noise=0.09, random_state=0)

# DBSCAN clustering
clusters, cluster_labels = DBSCAN(TEST, minpts=5, epsilon=0.3)

#Print cluster labels
#Cluster labels are assigned as follows: -1 for noise, 0 for border, 1, 2, 3, ... for core points.
print("Cluster labels:")
print(cluster_labels)
```

I am comparing my output with the output of DBSCAN from scikit-learn. The output variable labels_ contains the cluster labels for each point in the dataset given to fit(). Noisy samples are given the label -1. Refer to (https://scikit-learn.org/stable/modules/generated/sklearn.cluster.DBSCAN.html) for specifics. Ths code will output TRUE if the outputs match.

```python
dbscan_sklearn = sklearnDBSCAN(eps=0.3, min_samples=5)
dbscan_sklearn.fit(TEST)
scikit_labels = dbscan_sklearn.labels_

print("Scikit-learn DBSCAN labels:")
print(scikit_labels)

print("Custom DBSCAN labels:")
print(cluster_labels)

# Check if the labels match
labels_match = np.array_equal(scikit_labels, cluster_labels)
print("Labels Match:", labels_match)
```

Plotting the custom and scikit implementation 

```python
import matplotlib.pyplot as plt

# Create a scatter plot for Scikit-learn DBSCAN
plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.scatter(TEST[:, 0], TEST[:, 1], c=scikit_labels, cmap='viridis')
plt.title("Scikit-learn DBSCAN")

# Create a scatter plot for Custom DBSCAN
plt.subplot(122)
plt.scatter(TEST[:, 0], TEST[:, 1], c=cluster_labels, cmap='viridis')
plt.title("Custom DBSCAN")

plt.show()
```
